# visualizations

# histogram

```python
plt.figure()
plt.hist(iris['SL'],color='blue',rwidth=0.9,density='true')
plt.title('visualization 1',fontsize=15)
plt.xticks()
plt.yticks()
```

# scatter plots

```python
plt.figure(figsize=(10,12))
plt.subplot(2,1,1)
plt.scatter(iris['SL'],iris['SW'],s=20,c='Green')
plt.title('Plot of SL vs SW', fontsize=20)
plt.xlabel('SL',fontsize=16)
plt.ylabel('SW',fontsize=16,rotation=0)

plt.subplot(2,1,2)
plt.scatter(iris['PL'],iris['PW'],s=20,c='Blue')
plt.title('Plot of PL vs PW', fontsize=20)
plt.xlabel('PL',fontsize=16)
plt.ylabel('PW',fontsize=16,rotation=0)
plt.tight_layout()
```

# Pair plot

```python
sns.pairplot(iris)
```

# Heat Map


```python
sns.heatmap(iris.corr(),annot=True)
plt.tickets()
```
